## Kickoff 2023-03-12: Matplotlib demystified. What's the deal with fig and ax?

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.rand(20)
```

#### When we draw a graph using `plt`

- A `Figure` object is generated, usually called `fig`. Think of this as the "canvas"  
- An `Axes` object is generated implicitly, usually called `ax`. Think of this as the individual graph, or "cell"

#### If you're just whipping up a quick EDA graph, you generally can avoid `fig` and `ax`, and just use `plt.plot()`
- reminder: Pandas objects also have the .plot() method!

```python
# if we just want to plot one graph, we can do it like this, using the .plot() method and modifying our plt object:
plt.plot(np.random.rand(20))
plt.title('Always Title Your Charts')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### We can, however, take more control and eplicitly write out the plot using `plt.subplots()` and separate out `fig`, and `ax`:

```python
fig, ax = plt.subplots() # explicitly draw the cell so we can get the fig and ax objects
ax.plot(np.random.rand(20))
ax.set_title('Numpy can generate a squiggly line')
ax.set_xlabel("Here's the x axis")
ax.set_ylabel("Here's the y axis")

plt.show()
print(type(fig))
print(type(ax))
```

#### `.subplots()` was optional for the single plot above, but if we want to do more than one plot one at a time, it's required. For a 2x2 grid we supply (rows,cols) arguments:

`fig, axes = plt.subplots(2,2)`

*The axes object is indexable within that set range.*

```python
fig, ax = plt.subplots(2,2)

fig.suptitle("Modify the `fig` object to get a main title") # modifying our "canvas" with `fig`
fig.tight_layout()

ax[0,1].plot(np.random.rand(200)) # modifying individual plots with `ax`
ax[0,1].set_title('Obligatory Title')
ax[0,1].set_xlabel("x axis")
ax[0,1].set_ylabel("y axis")
fig.tight_layout()
```

#### Using a loop to make four plots at once:

```python
n_rows = 2
n_cols = 2

fig, axes = plt.subplots(n_rows, n_cols)
for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num] # which axes are we talking here within our 2,2 grid
        ax.plot(np.random.rand(20))
        ax.set_title(f'Plot ({row_num+1}, {col_num+1})')

fig.suptitle('Main title')
fig.tight_layout()
plt.show()
```

#### Alternate approach where instead of breaking out `fig` and `ax` we can just use `plt.subplot(rows,cols,cell_num)` to create each one.
- You can still modify `ax` 

```python
plt.subplots(figsize=(10,10))
plt.suptitle('main title')

for plt_num in range(1,5):
    plt.subplot(2,2,plt_num) # which one?
    plt.plot(np.random.rand(20)) # plot data
    plt.title(f'Plot {plt_num}') 

plt.tight_layout()
plt.show()
```
